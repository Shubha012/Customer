### **Co-relation Matrix with variable**

```python
sns.heatmap(data.corr())
```

```python
plt.figure(figsize=(10,10))
sns.barplot(x = "Profession" , y = "Age" , data = data , palette = "Spectral")
plt.title("Profession VS Age" , fontsize = 25 )
plt.xlabel("Profession", fontsize = 20)
plt.ylabel("Age" , fontsize = 15)
plt.show()
```

```python
sns.countplot(x=data["Gender"] )
plt.title("Number of MALE and FEMALE" , fontsize = 25 )
plt.xlabel("Gender", fontsize = 20)
plt.ylabel("Count" , fontsize = 15)
plt.show()
```

```python
plt.figure(figsize =(10,10))
sns.scatterplot(x = "Profession" , y = "Annual Income ($)" , hue = "Age" , data = data , palette = "flare")
plt.title("Profession VS Annual Income ($)" , fontsize=25)
plt.xlabel("Profession" , fontsize = 15)
plt.ylabel("Annual Income($)" , fontsize = 10)
plt.show()
```

```python
# Highest Annual Income which profession have?
plt.figure(figsize = (12,7))
sns.barplot(x = "Profession" , y = "Annual Income ($)" , data = data)
plt.title("Highest Income" , fontsize = 20)
plt.xlabel("Profession")
plt.ylabel("Annual Income ($)")
plt.show() 
```

```python
#Spending score of each group
plt.figure(figsize = (10,7))
sns.barplot(x = "Gender" , y = "Spending Score (1-100)" , data = data)
plt.title("Spending score " , fontsize = 20)
plt.xlabel("Gender")
plt.ylabel("Spending Score (1-100)")
plt.show() 
```

```python
plt.figure(figsize = (12,7))
sns.catplot(x = "Gender" , y = "Spending Score (1-100)" , data = data)
plt.title("Spending Score" , fontsize = 20)
plt.xlabel("Gender")
plt.ylabel("Spending Score (1-100)")
plt.show() 
```
